```python
import fitz
import pandas as pd

def extract_table(doc, page):
    # ページを取得する
    page = doc[page]

    # ページ上にあるテーブルを検出する
    tabs = page.find_tables()

    # 少なくとも1つのテーブルが見つかった場合
    table = []
    if tabs:
        # 最初のテーブルの内容を表示する
        table += tabs[0].extract()
    return table

def process_all_pages(pdf_path):
    # ドキュメントを開く
    doc = fitz.open(pdf_path)

    # すべてのページからテーブルを抽出する
    all_data = []
    for page_number in range(len(doc)):
        table = extract_table(doc, page_number)
        if table:
            all_data.extend(table)

    # DataFrameに変換する
    df = pd.DataFrame(all_data)
    return df

# PDFファイルのパス
pdf_path = "/content/drive/MyDrive/プレゼミ/economics_jikanwari.pdf"

# 全ページのテーブルを抽出して1つのDataFrameにする
df = process_all_pages(pdf_path)

# 結果を表示する
pd.DataFrame(df)
```

```python
df = df.fillna(method='ffill')
```

```python
df.columns = ['曜日', '講時', '開講期','講義コード','授業科目名','代表教員名','教室','備考']
```

```python
df=df.drop(df.index[0],)
```

```python
# すべての文字列列に対して\nを削除
df = df.apply(lambda x: x.str.replace('\n', '', regex=False) if x.dtype == "object" else x)
```

```python
df['授業科目名 (教室)'] = df['授業科目名'] + ' (' + df['教室'] + ')'
```

```python
from google.colab.files import upload_file
import streamlit as st
import spacy
import pandas as pd

#input
df = pd.read_csv('table_economics_jikanwari.csv')

#process
# ユニークな曜日、講時、開講期を取得
unique_days = df['曜日'].unique()
unique_periods = df['講時'].unique()
unique_terms = df['開講期'].unique()

# サイドバーで曜日を選択
selected_day = st.sidebar.selectbox('曜日を選択してください', unique_days)
# 「集中」の場合は自動的に講時を5に設定
if selected_day == '集中':
    selected_period = 5
else:
    selected_period = st.sidebar.selectbox('講時を選択してください', unique_periods)

# サイドバーで開講期を選択
selected_term = st.sidebar.selectbox('開講期を選択してください', unique_terms)

# 選択された条件に基づいて授業名をフィルタリング
filtered_classes = df[
    (df['曜日'] == selected_day) &
    (df['講時'] == selected_period) &
    (df['開講期'] == selected_term)
]
# フィルタリングされた授業の選択
if not filtered_classes.empty:
    class_options = ['-'] + filtered_classes['授業科目名 (教室)'].tolist()
    selected_class = st.selectbox('授業を選択してください', class_options)
else:
    st.write("該当する授業がありません。")
    selected_class = None



# 時間割表の初期化
if 'timetable' not in st.session_state:
    st.session_state.timetable = pd.DataFrame(index=unique_periods, columns=unique_days).fillna('')
# 選択された授業を時間割表に追加
if selected_class != '-':
  st.session_state.timetable.loc[selected_period, selected_day] = selected_class
elif selected_class == '-':
  st.session_state.timetable.loc[selected_period, selected_day] = ''


#output
st.write(f'曜日: {selected_day}, 講時: {selected_period}, 開講期: {selected_term} の授業名・教室・対象学部:')
st.write(filtered_classes[['授業科目名', '教室','備考']])
st.write("時間割表:")
st.data_editor(st.session_state.timetable, use_container_width=True)
#st.dataframe(st.session_state.timetable)
```
