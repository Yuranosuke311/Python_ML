### 並び

- `sort()`と`argsort()`

1. `sort()`は数値を**小さい順番（昇順）**から並べて出力する関数

1. `argsort()`は昇順に数値のインデックス(index)を出力する関数です。

```python
import numpy as np

name=np.array(["A","B","C"])
test = np.array([3,1,2])

print(np.sort(test)) # 数値の順番
print(np.argsort(test)) # 小 -> 大 のインデックス=1番小さな数字は左から1番目、2番目に小さい数字は2番目にある
print(name[np.argsort(test)])
```

降順に並びたい際には、最後に`[::-1]`を付けます。

```python
print(np.sort(test)[::-1])
print(np.argsort(test)[::-1])
print(name[np.argsort(test)[::-1]])
```

並べた配列を保存したい場合は、変数への代入を忘れずに

```python
test2 = np.sort(test)[::-1]
print(test2)
```

$$

$$



★チャレンジ★

二つ同じ次元の行列`A, B`が与えられたとき、各行における二つのベクトルのなす角(A[0]とB[0], A[1]とB[1], ....)を高い順番から（ベクトルa,ベクトルb, なす角)の形でプリントしてください。

```python
# コサインから角度を計算する関数
def cos2angle(cos):
    rad = np.arccos(cos)
    theta = rad * 180 / np.pi
    return theta

cos2angle(0.9)
```

```python
#AIの回答
import numpy as np

def angle_between_vectors(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    cos_theta = dot_product / (norm_a * norm_b)
    angle_radians = np.arccos(np.clip(cos_theta, -1.0, 1.0))
    angle_degrees = np.degrees(angle_radians)
    return angle_degrees

def print_sorted_angles(A, B):
    angles = []
    for vec_a, vec_b in zip(A, B):
        angle = angle_between_vectors(vec_a, vec_b)
        angles.append((vec_a, vec_b, angle))
    angles.sort(key=lambda x: x[2], reverse=True)
    for vec_a, vec_b, angle in angles:
        print(f"ベクトルa: {vec_a}, ベクトルb: {vec_b}, なす角: {angle}度")

# 例として、ランダムな行列AとBを生成してプリントします
A = np.random.rand(5, 3)  # 5行3列のランダムな行列
B = np.random.rand(5, 3)  # 5行3列のランダムな行列
print("与えられた行列A:")
print(A)
print("与えられた行列B:")
print(B)
print("各行のベクトル間の角度（高い順）:")
print_sorted_angles(A, B)
```

###<font color='crimson'>注意点(copy)</font>

pythonでは、リストを直接代入すれば、**参照渡し**になり、詳細は省きますが、両リストは連動するようになります。

```python
a = [1,2,3]
b = a # 参照渡し
print(b)

a[0] = 3
print('変更後')
print(a)
print(b)
```

この例では, bは弄ってないにもかかわらず、aと同じく最初の要素が3になりました。これを避けれために、pythonでは**copy()**を使い、**参照渡し**ではなく、元のオブジェクトをコピーして使います。

```python
import copy
a = [1,2,3]
b = copy.copy(a)
print(b)

a[0] = 3
print('変更後')
print(a)
print(b)
```

numpyでは、直接配列の後ろに`.copy()`を加えて使えます。

```python
a = np.array([1,2,3])
b = a.copy() #コピー
c = a # 参照渡し
print(b)

a[0] = 3
print('変更後')
print(a)
print(b)
print(c)
```

例外として、元となる配列が全体的に変わる時に、参照先の配列は自動的にコピーされ、元のオブジェクトとの関連が切られます。

```python
a = np.array([1,2,3])
b = a

a = np.array([3,2,1])
print(b) # bは変化してない
```
