##1.4 <font color='crimson'>ループ(Loop)</font>

プログラミングの強みの一つである、繰り返し処理について紹介します。

今後、いろんな計算をする際に、繰り返し計算をする場合が多いですが、その際に`for`分を用いた繰り返し構造を利用すれば数段楽になります。

###<font color='crimson'>range</font>

まず range()について説明します。


```python
range(5)
```

リストに変換すれば

```python
list(range(5))
```

すなわち、 range(3)であれば[0,3)の範囲の整数が生成されます。

ここで、たとえば、「0から5までの整数をプリントしてください。」の課題を考えます。

```python
print(0)
print(1)
print(2)
print(3)
print(4)
print(5)
```

５までならなんとかかけますが、「0から100までの整数をプリントしてください。」はさすがに非効率的になります。そこで for文の登場です

```python
for i in range(6):
  print(i)
```

```python
for i in range(6):
  print((i+1)*2)
```

これを解析していきます。まず一行のforは繰り返し文の宣言です。続いて、最後のrange(6)は、繰り返す主体です。また、iが登場しますが、これは毎回の繰り返す数値を代入する変数になります。

たとえば、range(6)の構成は[0, 1, 2, 3, 4, 5]であり、最初の一回の繰り返しに最初の要素である0を、iに代入します。そこでi=0になり、print(i)で0をプリントします。同じ原理で、次のループではi=1で、0から順番にプリントしていく仕組みです。


このforを扱えば、0から100までの整数を余裕でプリントできます。

```python
for x in range(100):
  print(x)
```

for文は、列を揃うことによって成立し、列を戻せば終了します。

```python
for i in range(5):
  b = i * 2  # ここからfor文の中身
  ###
  ###
  print(b)   #　ここまでがfor文

print('for文がおわりました')
```

for文は複数適応することも可能です。

```python
for i in range(3):
  for j in range(3):
    print(i, '*', j, '=', i*j)
```

また、range()のみならず、リスト型も扱えます。

```python
a = [1, 'hello', 2]
for i in a:
  print(i)
```

##1.5 <font color='crimson'>条件分岐(if)</font>

プログラミングのもう一つの重要な要素として、条件分岐が挙げられます。
条件分岐は、ある条件を満たしているのみ、コードを実行するという仕組みになっています。利用には`if`文を使います。

```python
a = 5
if a > 3:
  print('この数字は3より大きいです。')

if a > 10:
  print('この数字は10より大きいです。')
```

比較の記号として、 >, >=, <, <=, ==などがあります。

```python
print('数字を入力してください。')
a = int(input())
if a == 3:
  print('この数字は3です。')

if a>3:
  print('この数字は3ではありません。')

if a<3:
  print('この数字は3ではありません。')
```

###<font color='crimson'>else, elif</font>

ここで、elseを入れることにより、条件を満たさない場合のコードを示すことができます。

```python
print('数字を入力してください。')
a = float(input())
if a == 3:
  print('この数字は3です。')
else:
  print('この数字は3ではありません。')
```

さらに、elifを使うと、複数の分岐が可能になります。

```python
print('数字を入力してください。')
a = float(input())

if a < 3:
  print('この数字は3より小さい')
elif a < 5:
  print('この数字は3と同じか大きく、5より小さい')
else:
  print('この数字は5と同じか大きい')
```

###<font color='crimson'>and, orとnot</font>

- not: 条件の反転を意味しています。

```python
print('数字を入力してください。')
a = float(input())

if not a >= 5: # if a < 5と同じ意味
  print('入力した数字は5より小さいです。')
```

- and: 複数の条件を全部満たす必要があります。

```python
print('数字を入力してください。')
a = float(input())

if a > 5 and a < 10:
  print('入力した数字は5より大きく、10より小さいです。')
```

- or:条件のいずれかを満足する必要があります。

```python
print('数字を入力してください。')
a = float(input())

if a > 5 or a < 3:
  print('入力した数字は5より大きいか、3より小さいです。')
```

```python
#チャレンジ4
import re
a = ['3', 'dog', '1', 'cat', '9.3', '6']
r = re.compile('^[0-9]+$') # 「最初から最後まで数字が1文字以上続く」という式
# re_sampleに該当する文字列だけを抽出した配列を生成し、resultに代入。
for b in filter(r.match,a):
  print(b)
```
