## プロンプトに対して音声（出力）

```python
##質問に音声で答える
import base64
from openai import OpenAI

client = OpenAI()

completion = client.chat.completions.create(
    model="gpt-4o-audio-preview",
    modalities=["text", "audio"],
    audio={"voice": "alloy", "format": "wav"},
    messages=[
        {
            "role": "user",
            "content": "Is a golden retriever a good family dog?"
        }
    ]
)

print(completion.choices[0])

wav_bytes = base64.b64decode(completion.choices[0].message.audio.data)
with open("dog.wav", "wb") as f:
    f.write(wav_bytes)
```

## 音声（入力）から回答生成

```python
import base64
import requests
from openai import OpenAI

client = OpenAI()

# Fetch the audio file and convert it to a base64 encoded string
audio_file = open("/content/drive/MyDrive/プログラミング（AI活用）/dog.wav","rb")
wav_data = audio_file.read()
##when url
#url = "https://openaiassets.blob.core.windows.net/$web/API/docs/audio/alloy.wav"
#response = requests.get(url)
#response.raise_for_status()
#wav_data = response.content

encoded_string = base64.b64encode(wav_data).decode('utf-8')

completion = client.chat.completions.create(
    model="gpt-4o-audio-preview",
    modalities=["text", "audio"],
    audio={"voice": "alloy", "format": "wav"},
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "What is in this recording?"
                },
                {
                    "type": "input_audio",
                    "input_audio": {
                        "data": encoded_string,
                        "format": "wav"
                    }
                }
            ]
        },
    ]
)

print(completion.choices[0].message.audio.transcript)#message内のaudioの中のtranscriptにある
```

## [TTS-model](https://platform.openai.com/docs/guides/text-to-speech)

・書かれたブログ投稿をナレーションする<br>
・複数の言語で音声を生成<br>
・ストリーミングを使用してリアルタイムのオーディオ出力を提供する

```python
##入力テキストから音声を生成
from pathlib import Path
from openai import OpenAI
client = OpenAI()

speech_file_path = Path("/content/drive/MyDrive/プログラミング（AI活用）/出力/speech_file(8-1)").parent / "speech.mp3"
response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",#echo,fable,onyx,nova,shimmer
  input="Today is a wonderful day to build something people love!"
)

response.stream_to_file(speech_file_path)#"speech.mp3"で保存される
```
